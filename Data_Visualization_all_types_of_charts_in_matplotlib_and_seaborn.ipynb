#Figure and Axes

```python
x=[2,3,4,5,6,7]
y=[10,20,30,40,50,60]
z= [100,67,45,78,100,67]
fig,ax = plt.subplots()  # create a figure and a single subplot (axes)
ax.plot(x, y, color = 'green', linestyle = ':', label = 'vs Y', marker = 'o')
ax.plot(x, z, color = 'blue', linestyle = '--', label = 'vs a', marker = '^')
ax.set_title('Basic Line Plot using matplotlib')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.legend()
plt.show()
```

```python
# fig represents the entire figure.
# ax (or axes) is the individual plotting area within the figure.
```

```python
#set the figure size
#We can specify the size (width, height in inches) by passing figsize:
fig,ax = plt.subplots(figsize = (3,3)) # create a figure and a single subplot (axes)
ax.plot(x, y, color = 'green', linestyle = ':', label = 'vs Y', marker = 'o')
ax.plot(x, z, color = 'blue', linestyle = '--', label = 'vs a', marker = '^')
ax.set_title('Basic Plotting using Matplotlib')
ax.set_xlabel('Height')
ax.set_ylabel('Data')
ax.legend()
plt.show()
fig.savefig("my_plot_2.png", dpi=300)
```

```python
# Add a super-title for the entire figure
# Use fig.suptitle() to set a title that appears above all subplots:
fig, ax = plt.subplots()
ax.plot(x, y)
fig.suptitle("My Entire Figure Title", fontsize=14) # Figure : Super Title
ax.set_title('Example Plot') # Plot Specific title
plt.show()
```

```python
# Tight layout
# If subplots or labels overlap, you can improve spacing:
# plt.subplots(2, 1): creating a figure with 2 rows and 1 column of subplots  i.e., two vertically stacked plots.

fig, ax = plt.subplots(2, 1, figsize=(6, 6))
ax[0].plot(x, y, color='green', linestyle='-.', label='vs Y', marker = '^')
ax[1].plot(x, z,marker = 'o' ,color='blue', linestyle='--', label='vs a')
#After executing the above two queries we can see that two charts are appearing
#one in [0]th index (top1) and another one is [1]th index (bottom one).
#Now we have to give labels and give title
ax[0].set_xlabel('Height')
ax[1].set_xlabel('Height')
ax[0].set_ylabel('Data')
ax[1].set_ylabel('Data')
ax[0].set_title('Data plot title for index 0')
ax[1].set_title('Data plot title for index 1')
#Also, we have to add supertitle
fig.suptitle('Plot Super Title')
#After executing this command my title has been overlapped
#so we will use tight layout function to adjust the space
fig.tight_layout()
#Now my plots are looking ok
plt.show()

```

```python
# plt.subplots(2, 2): creating a figure with 2 rows and 2 columns of subplots  four plots.
x=[2,3,4,5,6,7]
y=[10,20,30,40,50,60]
z= [100,67,45,78,100,67]
a = [45,46,89,100,23,1]

fig, ax = plt.subplots(2, 2, figsize=(8, 8))

ax[0,0].plot(x, y, color='green', linestyle='-.', label='vs Y', marker = '^')
ax[0,1].plot(x, z,marker = 'o' ,color='blue', linestyle='--', label='vs a')
ax[1,0].plot(y, z,marker = '^' ,color='red', linestyle='-.', label='vs a')
ax[1,1].plot(y, a,marker = '^' ,color='orange', linestyle='-.', label='vs a')

ax[0,0].set_xlabel('Height')
ax[0,1].set_xlabel('Height')
ax[1,0].set_xlabel('Height')
ax[1,1].set_xlabel('Height')

ax[0,0].set_ylabel('Data')
ax[0,1].set_ylabel('Data')
ax[1,0].set_ylabel('Data')
ax[1,1].set_ylabel('Data')

ax[0,0].set_title('Data plot title for index 0,0')
ax[0,1].set_title('Data plot title for index 0,1')
ax[1,0].set_title('Data plot title for index 1,0')
ax[1,1].set_title('Data plot title for index 1,1')

#Also, we have to add supertitle
fig.suptitle('Plot Super Title')

#After executing this command my title has been overlapped
#so we will use tight layout function to adjust the space
fig.tight_layout()
#Now my plots are looking ok

plt.show()

#If i want to save my plot
fig.savefig('my_plot.png', dpi =300)
#dpi= resolution
```

```python
# Add text anywhere on the figure
# Use fig.text(x, y, s, ...) to
# place text in figure coordinates (where 0,0 is bottom-left and 1,1 is top-right):
fig, ax = plt.subplots()
ax.plot(x, y)

# x=0.5, y=0.02 means horizontally centered, near the bottom
fig.text(0.5, 0.1, 'Figure-wide X-axis Label', ha='center', fontsize=12)
# 0.5 : Middle of X Axis : 50% of overall screen
# 0.1 : 10% of y Axis :
plt.show()
# ha='center' : generally it's by default center | Aligns the text in the position (left, right, center)
```

#Histogram

```python
# Histogram | Analyse the distribution of Single Continuous Variable | Numerical variable : Age, Salary, Experience
x = [5,4,10,15,24,14,16,20,36,23,10,45,34,70,69,45,67,90,35,23,56,67,89,54] # Points or marks scored by player
plt.hist(x, bins=10, color='green',edgecolor='white') # bins = # of buckets to be created
plt.title('Histogram Example')
plt.show()
plt.tight_layout()
# Buckets - [0-10), [10,20).....
# For each buckets, how many data points are present
# [0-10) ---> 2 | [ --> Inclusive , ) --> Exclusive
# [10,20) --> 5
# [20,30) -->
# (20,40] -->
# Distribution -->
# UseCases :
# Apply some transformation in ML for a particular column to make it normal distribution
# Outlier Detection
# Help you estimate the data for missing values imputation | mean, median
```

```python
# Create a Matplotlib figure with subplots
fig, axes = plt.subplots(1, 3, figsize=(10, 5))  # 1 row, 3 columns

# Subplot 1: Scatterplot
sns.scatterplot(x='total_bill', y='tip', data=tips, ax=axes[0], color = 'green', hue = 'sex')
axes[0].set_title('Scatterplot: Total Bill vs Tip')
axes[0].set_xlabel('Total Bill ($)')
axes[0].set_ylabel('Tip ($)')

# Subplot 2: Boxplot
sns.boxplot(x='day', y='total_bill', data=tips, ax=axes[1])
axes[1].set_title('Boxplot: Total Bill by Day')
axes[1].set_xlabel('Day of the Week')
axes[1].set_ylabel('Total Bill ($)')

sns.kdeplot(data=tips['total_bill'], ax = axes[2])
axes[2].set_title('Total Bill KDE Plot')
axes[2].set_xlabel('Day of the Week')
axes[2].set_ylabel('Total Bill ($)')

# Adjust layout and show the plots
plt.tight_layout()
plt.show()
```
