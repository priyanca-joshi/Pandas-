## .merge()

```python
'''
pd.merge(left, right, how='inner', on=None, left_on=None, right_on=None)
Parameters:
left, right: DataFrames to merge

how: Type of join – 'inner', 'outer', 'left', 'right'

on: Column(s) to join on (must be common in both)

left_on, right_on: Use when column names differ

suffixes: Add suffixes to overlapping column names
'''
```

```python
# key --> Common information/ Id across tables/Dataframe
```

```python
# INNER JOIN (default)

df1 = pd.DataFrame({'ID': [1, 2, 3], 'Name': ['Alice', 'Bob', 'Charlie']})
df2 = pd.DataFrame({'ID': [2, 3, 4], 'Marks': [85, 90, 75]})
```

```python
df1
```

```python
df2
```

```python
# Common Information across df1 & df2 --> key --> ID
# if key is not present | merging can't be done
```

```python
result = pd.merge(df1,df2, on='ID')# on --> key | common info
print(result)
```

```python
# Criterias of Joining
# inner, left, right, outer
# inner --> only keep common data between 2 dataframes
```

```python
result_inner = pd.merge(df1,df2, on='ID', how='inner')
print(result_inner)
```

# Left join

```python
# Left join --> as the criteria for Joining
# All data of left dataframe and matching ones for right dataframe
```

```python
df1
```

```python
df2
```

```python
result_left = pd.merge(df1,df2, on='ID', how='left')
print(result_left)
```

```python
result_left = pd.merge(df2,df1, on='ID', how='left')
print(result_left)
```

```python
# df1 --> present in left side --> all data/keys of this
# df2 --> present in right side --> only common ones
```

# Right join

```python
# RIGHT JOIN --> opposite of left join
#  all of right and matching for left
# df2 --> right --> all ids of df2
# df1 --> left --> only matching ones
```

right is not used frequently because it's easy to switch places of dataframe

```python
df1
```

```python
df2
```

```python
right_join = pd.merge(df1,df2, on='ID', how = 'right')
print(right_join)
```

```python
right_joinn= pd.merge(df2,df1, on = 'ID', how = 'right')
print(right_joinn)
```

#Outer Join

```python
# OUTER JOIN
# Full join --> left + right
```

```python
df1
```

```python
df2
```

```python
result_outer = pd.merge(df1, df2, on = 'ID', how = 'outer')
print(result_outer)
# df1 & df2 --> all information as key
```

```python
# Concat and Outer is not same
# Concat --> no concept of key | dataframe should have similar information
# Outer --> concept of key | dataframe can have different information with keys as common criteria to join
```

```python
# key & Criteria --> Important concept
# key --> Common Information
# Critria --> inner, left, right or outer | depends upon data to be added for analysis
```

```python
# Merge on Different Column Names

df1 = pd.DataFrame({'emp_id': [1, 2], 'Name': ['Alice', 'Bob']})
df2 = pd.DataFrame({'id': [1, 2], 'Salary': [50000, 60000]})

# key is id but name of the column is different
# Change the column nameemp_id --> id and on --> 'id'
```

```python
df1
```

```python
df2
```

```python
result = pd.merge(df1, df2, left_on='emp_id', right_on='id', how='inner')
print(result)
# name of column as key in left table --> left_on
# name of column as key in right table --> right_on
```

```python
# Merging When Keys Have Duplicates
# When one or both DataFrames have duplicate values in the key column,
# the result will be all combinations of matching rows.
```
